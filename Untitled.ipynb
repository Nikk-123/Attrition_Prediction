```python
import pandas as pd

# Load the dataset
df = pd.read_csv('indian_customer_data.csv')

# Display the first few rows
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Handle missing values
df.dropna(inplace=True)

# Encode categorical variables
label_encoders = {}
for col in ['Gender', 'City', 'State', 'Types of Products']:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# Display the first few rows after preprocessing
print(df.head())
```

```python
# Features and target variable
X = df.drop(['Customer ID', 'Name', 'Attrition Risk'], axis=1)
y = df['Attrition Risk'].map({'Yes': 1, 'No': 0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.preprocessing import StandardScaler

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Use the scaled data to train
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train the model
logreg_scaled = LogisticRegression(max_iter=1000)
logreg_scaled.fit(X_train_scaled, y_train)

# Evaluate
y_pred_scaled = logreg_scaled.predict(X_test_scaled)
print(classification_report(y_test, y_pred_scaled))


# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_scaled)
conf_matrix = confusion_matrix(y_test, y_pred_scaled ) 
class_report = classification_report(y_test, y_pred_scaled)

print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{conf_matrix}")
print(f"Classification Report:\n{class_report}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay, precision_recall_curve

# Confusion Matrix Plot
ConfusionMatrixDisplay.from_estimator(logreg_scaled, X_test_scaled, y_test, cmap='Blues')
plt.title('Confusion Matrix')
plt.show()

# Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, logreg_scaled .predict_proba(X_test_scaled)[:, 1])
plt.plot(recall, precision, color='purple')
plt.title('Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```

```python
# Feature importances from logistic regression
coef = logreg_scaled.coef_[0]
features = X.columns

# Plot feature importances
plt.figure(figsize=(10, 6))

# Assuming coef and features are defined
sns.barplot(x=coef, y=features, hue=features, palette='viridis', legend=False)
plt.title('Feature Importance')
plt.xlabel('Coefficient Value')
plt.ylabel('Feature')
plt.show()
```
